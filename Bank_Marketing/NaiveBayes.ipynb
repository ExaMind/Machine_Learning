```python
import numpy as np
import pandas as pd
from sklearn import preprocessing, cross_validation
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv('../data/bank-additional-full.csv')

df
```

```python
number = LabelEncoder()
age = number.fit_transform(df['job'].astype('str'))

df['job'] = age

marital = number.fit_transform(df['marital'].astype('str'))
df['marital'] = marital

education = number.fit_transform(df['education'].astype('str'))
df['education'] = education

default = number.fit_transform(df['default'].astype('str'))
df['default'] = default

housing = number.fit_transform(df['housing'].astype('str'))
df['housing'] = housing

loan = number.fit_transform(df['loan'].astype('str'))
df['loan'] = loan

contact = number.fit_transform(df['contact'].astype('str'))
df['contact'] = contact

month = number.fit_transform(df['month'].astype('str'))
df['month'] = month

day_of_week = number.fit_transform(df['day_of_week'].astype('str'))
df['day_of_week'] = day_of_week

poutcome = number.fit_transform(df['poutcome'].astype('str'))
df['poutcome'] = poutcome

df
```

```python
X = np.array(df.drop(['y'], 1))
y = np.array(df['y'])

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3)

clf = GaussianNB()
clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)
print(accuracy)
```
